### Visualization

```python
plt.figure(figsize = (6,6))
plt.title('Ship Mode')
plt.pie(ds['Ship Mode'].value_counts(), labels=ds['Ship Mode'].value_counts().index,autopct='%1.1f%%')
plt.show()
```

```python
plt.figure(figsize = (6,6))
plt.title('Segement')
plt.pie(ds['Segment'].value_counts(), labels=ds['Segment'].value_counts().index,autopct='%1.1f%%')
plt.show()
```

```python
plt.figure(figsize = (6,6))
plt.title('Region')
plt.pie(ds['Region'].value_counts(), labels=ds['Region'].value_counts().index,autopct='%1.1f%%')
plt.show()
```

```python
plt.figure(figsize = (6,6))
plt.title('Category')
plt.pie(ds['Category'].value_counts(), labels=ds['Category'].value_counts().index,autopct='%1.1f%%')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
ds.groupby('Ship Mode')['Profit'].sum().sort_values().plot.bar()
plt.title('Profits on Different Ship Modes')
```

```python
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
ds.groupby('Segment')['Profit'].sum().sort_values().plot.bar()
plt.title('Profits on Different Segments')
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
ds.groupby('Country')['Profit'].sum().sort_values().plot.bar()
plt.title('Profits in Different Countries')
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
ds.groupby('Region')['Profit'].sum().sort_values().plot.bar()
plt.title('Profits in Different Regions')
```

```python
plt.figure(figsize=(15,5))
ds.groupby('State')['Profit'].sum().sort_values().plot.bar()
plt.title('Total Profits in Different States')
plt.show()
```

```python
top_subcategory_s = ds.groupby("Sub-Category").Sales.sum().nlargest(n = 100)
# computing top sub-categories in terms of profit from first 100 observations
top_subcategory_p = ds.groupby("Sub-Category").Profit.sum().nlargest(n = 100)

# plotting to see it visually
plt.style.use('seaborn')
top_subcategory_s.plot(kind = 'bar',figsize = (10,5),fontsize = 14)
top_subcategory_p.plot(kind = 'bar',figsize = (10,5),fontsize = 14, color = 'red')
plt.xlabel('Sub-Category',fontsize = 15)
plt.ylabel('Total Sales/Profits',fontsize = 15)
plt.title("Top Sub-Category Sales vs Profit",fontsize = 15)
plt.show()
```

```python
plt.figure(figsize=(14,12))
statewise = ds.groupby(['City'])['Profit'].sum().nlargest(50)
statewise.plot.barh() 
```

```python
plt.figure(figsize=(10,5))
statewise = ds.groupby(['Discount'])['Profit'].sum().nlargest(50)
statewise.plot.barh() 
```

```python
plt.figure(figsize = (6,6))
plt.title('Discount')
plt.pie(ds['Discount'].value_counts(), labels=ds['Discount'].value_counts().index,autopct='%1.1f%%')
plt.show()
```

```python
sns.catplot(x="Region", y="Profit", kind="bar",hue='Ship Mode', data=ds)
plt.title('Profits on Segment with different Categories ')
plt.show()
```

```python
sns.catplot(x="Region", y="Profit", kind="bar",hue='Category', data=ds)
plt.title('Profits on Segment with different Categories ')
plt.show()
```

```python
sns.catplot(x="Segment", y="Profit", kind="bar",hue='Category', data=ds)
plt.title('Profits on Segment with different Categories ')
plt.show()
```
